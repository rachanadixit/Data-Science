```python
import numpy as np
import matplotlib.pyplot as plt

data = {'Data Structure':55, 'Data Science':68, 'OOPS':69,
        'BSE':54}
courses = list(data.keys())
values = list(data.values())
 
fig = plt.figure(figsize = (5, 2))


plt.bar(courses, values, color =["green","yellow","blue","red"],
        width = 0.2)

plt.xlabel("Courses")
plt.ylabel("No. of students Passed ")

plt.show()
```

```python
import numpy as np
array1 = np.array([5,10,15,20])
array2 = np.array([2, 4, 6, 8])

addition = array1 + array2
subtraction = array1 - array2
multiplication = array1 * array2
division = array1 / array2

print("Addition Result:", addition)
print("Subtraction Result:", subtraction)
print("Multiplication Result:", multiplication)
print("Division Result:", division)
```

```python
import numpy as np

A = np.array([[1, 2], 
              [3, 4]])
B = np.array([[5, 6], 
              [7, 8]])
C = np.matmul(A, B)
D = np.add(A,B)

print("Result of matrix multiplication:\n", C)
print("Result of matrix additom,:\n", D)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0,50, 100)  
y = np.sin(x)                

plt.plot(x, y, label='Sine Wave', color='green')


plt.title('Simple Line Plot of Sine Wave')
plt.xlabel('X-axis (radians)')
plt.ylabel('Y-axis (amplitude)')
plt.legend()
plt.grid(True)


plt.show()
```
